# Değişken Dönüşümleri

```python
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('tips')
df.head()
```

# 0-1 Dönüşümü

```python
from sklearn.preprocessing import LabelEncoder

lbe = LabelEncoder()
```

```python
lbe.fit_transform(df['sex'])
```

```python
df['yeni_sex'] = lbe.fit_transform(df['sex'])
df.head()
```

# "1 ve Diğerleri (0)" Dönüşümü

```python
# belirttiğimiz değişkene 1 diğerlerine 0 atar.
df.head()
```

```python
import numpy as np
df['yeni_day'] = np.where(df['day'].str.contains('Sun'), 1, 0)
```

```python
df
```

# Çok Sınıflı Dönüşüm

```python
from sklearn.preprocessing import LabelEncoder
lbe = LabelEncoder()
```

```python
lbe.fit_transform(df['day'])
```

# One-Hot Dönüşümü ve Dummy Değişken Tuzağı

```python
df.head()
```

```python
# yeni oluşturulan 2 değişken birbirini tamamlıyor. Fakat birisi olmadan diğeri anlaşılabilir. buna dummy tuzağı denilir.
df_one_hot = pd.get_dummies(df, columns=['sex'], prefix=['sex'])
df_one_hot.head()
```
