# Eksik Veri Silme İşlemleri

```python
import pandas as pd
import numpy as np
```

```python
V1 = np.array([1,3,6,np.NaN,7,1,np.NaN,9,15])
V2 = np.array([7,np.NaN,5,8,12,np.NaN,np.NaN,2,3])
V3 = np.array([np.NaN,12,5,6,14,7,np.NaN,2,31])

df = pd.DataFrame({
    'V1': V1,
    'V2': V2,
    'V3': V3
})
df
```

```python
df.dropna()
# en az 1 eksik değeri olan gözlemler silinir.
```

```python
df.dropna(how='all')
# Sadece hepsi eksik olan gözlemler silinir.
```

```python
df.dropna(axis=1)
# değişken bazında bakıldığında en az 1 tane eksik değeri olan gözlemi sil.
```

```python
df['sil_beni'] = np.NaN
```

```python
df.dropna(axis=1,  how='all', inplace=True)
df
```
