# Değer Atama Yöntemleri

```python
import pandas as pd
import numpy as np
```

```python
V1 = np.array([1,3,6,np.NaN,7,1,np.NaN,9,15])
V2 = np.array([7,np.NaN,5,8,12,np.NaN,np.NaN,2,3])
V3 = np.array([np.NaN,12,5,6,14,7,np.NaN,2,31])

df = pd.DataFrame({
    'V1': V1,
    'V2': V2,
    'V3': V3
})
df
```

# Sayısal Değişkenlerde Atama

```python
df['V1'].fillna(0)
# eksik değerlere 0 atandı.
```

```python
df['V1'].fillna(df['V1'].mean())
# eksik değerlere o değişkenin ortalama değeri atandı.
```

```python
df.apply(lambda x: x.fillna(x.mean()))
# tüm sayısal değişkenlerde eksik değerlerin yerine o değişkenin ortalaması atandı.
```

```python
# sayısal değişkenlerde eksik değerlerin yerine o değişkenin ortalamasını atama
# 2. yol

df.fillna(df.mean()[:])
```

### V1 ve V2 değişkenlerini kendi ortalamaları ile, V3 değişkenini ise kendi medianı ile doldurmak istediğimizi düşünelim.

```python
df.fillna(df.mean()['V1':'V2'], inplace=True)
```

```python
df['V3'].fillna(df['V3'].median(), inplace=True)
```
